```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import os

# Load results
df = pd.read_csv("results/captions/auto_tuned_batch_results.csv")

# ðŸ“ˆ 1. Plot ISP vs BERTScore
plt.figure(figsize=(12, 5))
for param in ["brightness", "contrast", "sharpness"]:
    plt.subplot(1, 3, ["brightness", "contrast", "sharpness"].index(param)+1)
    sns.boxplot(x=param, y="bert", data=df)
    plt.title(f"{param.capitalize()} vs BERTScore")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
